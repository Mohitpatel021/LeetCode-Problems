```python
    #Question:240 Search a 2D Matrix II
    #Link :https://leetcode.com/problems/search-a-2d-matrix-ii/description/

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        r,c=len(matrix)-1,0
        while r>=0 and c<len(matrix[0]):
            if matrix[r][c]==target:
                print( "If", matrix[r][c],r,c)
                return True
            elif matrix[r][c]<target:
                print("elif",matrix[r][c],r,c)
                c+=1
            else:
                print("else",matrix[r][c],r,c)
                r-=1
        return False
```

```python
    #Question:885. Spiral Matrix III
    #Link :https://leetcode.com/problems/spiral-matrix-iii/description/



class Solution:
    def spiralMatrixIII(self, rows: int, cols: int, rStart: int, cStart: int) -> List[List[int]]:
        position = [rStart, cStart]
        movements = [[0,1],[1,0],[0,-1],[-1,0]]
        res = []
        i = 0
        while len(res) < rows*cols:
            if -1 < position[0] < rows and -1 < position[1] < cols:
                res.append(position[:])
            position[0] += movements[i%4][0]
            position[1] += movements[i%4][1]
            
            i += 1
            if [position[0] + movements[i%4][0], position[1] + movements[i%4][1]] in res:
                i -=1
            
        return res
```

```python
    #Question:1861. Rotating the Box
    #Link :https://leetcode.com/problems/rotating-the-box/description/

class Solution:
    def rotateTheBox(self, boxGrid: List[List[str]]) -> List[List[str]]:
        for row in boxGrid:
            left = right = len(row) - 1
            while left >= 0:
                if row[left] == '*':
                    left -= 1
                    right = left
                
                elif row[left] == '#':
                    row[right] = row[left]
                    if left != right: row[left] = '.'
                    right -= 1
                    left -= 1
                
                else:
                    left -= 1
        
        answer = []
        for col in range(len(boxGrid[0])):
            curr = []
            for row in range(len(boxGrid) - 1, -1, -1):
                curr.append(boxGrid[row][col])
            answer.append(curr)
        
        return answer
            
```

```python
  #Question:1975. Maximum Matrix Sum
    #Link :https://leetcode.com/problems/maximum-matrix-sum/description/


class Solution:
    def maxMatrixSum(self, matrix: list[list[int]]) -> int:
        tot, negCnt = 0, 0
        zeroExist = False
        negMin = float('inf')
        for row in matrix:
            for x in row:
                if x < 0:
                    negCnt += 1
                if x == 0:
                    zeroExist = True
                tot += abs(x)
                negMin = min(negMin, abs(x))
        if negCnt % 2 == 0 or zeroExist:
            return tot
        return tot - 2 * negMin
```
